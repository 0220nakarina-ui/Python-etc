```python
#練習
#リスト化・辞書化克服用
#花束を作ろう
#使いたい花→紫陽花(青)・紫リンドウ・紫トルコキキョウ・レモンリーフ
list_flower = [
       {"flower":"紫陽花(青)","category":"花","price":"1000"},
       {"flower":"紫リンドウ","category":"花","price":"250"},
       {"flower":"紫トルコキキョウ","category":"花","price":"600"},
       {"flower":"レモンリーフ","category":"葉","price":"300"},
]

#紫陽花(青)・紫リンドウ・紫トルコキキョウ・レモンリーフで合計\5000以内で作りたい
#紫陽花(青)メインという内容のオーダーが入ったよ、作ろう

budget = 5000            #\5000以内
num_flowers = 5          #花の合計数・自由に変えられる
bouquet = []             #花がまだ無いよ
total = 0                #まだ0円

#紫陽花(青)がメイン
for _ in range(num_flowers):
    price = int(list_flower[0]["place"])
    if total + price <= budget:
        bouquet.append(list_flower[0]["flower"])
        total += price

#他の花も入れよう

```
