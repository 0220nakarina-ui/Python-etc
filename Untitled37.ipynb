```python
#練習
#リスト化・辞書化克服用
#花束を作ろう
#使いたい花→紫陽花(青)・紫リンドウ・紫トルコキキョウ・レモンリーフ
list_flower = [
       {"flower":"紫陽花(青)","category":"花","price":"1000"},
       {"flower":"紫リンドウ","category":"花","price":"250"},
       {"flower":"紫トルコキキョウ","category":"花","price":"600"},
       {"flower":"レモンリーフ","category":"葉","price":"300"},
]

#紫陽花(青)・紫リンドウ・紫トルコキキョウ・レモンリーフで合計\5000以内で作りたい
#紫陽花(青)メインという内容のオーダーが入ったよ、作ろう

budget = 5000            #\5000以内
num_flowers = 5          #花の合計数・自由に変えられる
bouquet = []             #花がまだ無いよ
total = 0                #まだ0円


#とりあえず動いたが思ってた挙動と全然違うし、最初のループがよく分かってない

#紫陽花(青)がメイン
for flower in list_flower:                    #for→何を基準に繰り返すか/変数→flower
    price = int(list_flower[0]["price"])      #[0]→常に紫陽花(青)/数字を変えると他の花になる
    if total + price <= budget:
        bouquet.append(list_flower[0]["flower"])
        total += price

# 他の花も入れよう
for index in range(1, len(list_flower)): #紫リンドウを使うよ/数字を変えると他の花になる
    flower = list_flower[index]
    name = flower["flower"]
    price = int(flower["price"])
    category = flower["category"]

#あとは予算に合わせてアレンジ
    if total + price <= budget:
        bouquet.append(name)
        total += price

print(bouquet)
print(total)
```
